**Import Statements**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Classes**

```python
x = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

**Precision Plot**

```python
y1 = [93, 94, 92, 81, 87, 88, 81, 94, 95, 89] # SequentialNetAug
y2 = [84, 85, 89, 80, 80, 80, 64, 89, 89, 83] # ResidualNetAug
y3 = [88, 88, 81, 72, 89, 61, 88, 85, 88, 87] # SqueezeNetAug
y4 = [91, 95, 89, 81, 87, 93, 80, 91, 90, 89] # DepthwiseConvNetAug
plt.plot(x, y1)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.scatter(x, y2)
plt.plot(x, y3)
plt.scatter(x, y3)
plt.plot(x, y4)
plt.scatter(x, y4)
plt.title('Classwise Precision Plot')
plt.ylabel('Precision')
plt.xlabel('Class')
plt.legend(['SequentialNetAug', 'ResidualNetAug', 'SqueezeNetAug', 'DepthwiseConvNetAug'], bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation = 60)
plt.ylim(bottom=0, top=100)
plt.show()
```

**Recall Plot**

```python
y1 = [88, 96, 82, 80, 90, 81, 97, 92, 92, 95] # SequentialNetAug
y2 = [84, 94, 64, 56, 82, 74, 96, 85, 90, 90] # ResidualNetAug
y3 = [81, 93, 79, 55, 73, 88, 87, 85, 90, 89] # SqueezeNetAug
y4 = [89, 95, 84, 76, 89, 70, 98, 93, 96, 95] # DepthwiseConvNetAug
plt.plot(x, y1)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.scatter(x, y2)
plt.plot(x, y3)
plt.scatter(x, y3)
plt.plot(x, y4)
plt.scatter(x, y4)
plt.title('Classwise Recall Plot')
plt.ylabel('Recall')
plt.xlabel('Class')
plt.legend(['SequentialNetAug', 'ResidualNetAug', 'SqueezeNetAug', 'DepthwiseConvNetAug'], bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation = 60)
plt.ylim(bottom=0, top=100)
plt.show()
```

**F1-Score Plot**

```python
y1 = [90, 95, 87, 80, 88, 85, 89, 93, 94, 92] # SequentialNetAug
y2 = [84, 91, 74, 66, 81, 77, 77, 87, 90, 86] # ResidualNetAug
y3 = [84, 91, 80, 62, 80, 72, 87, 85, 89, 88] # SqueezeNetAug
y4 = [90, 95, 86, 78, 88, 80, 88, 92, 93, 92] # DepthwiseConvNetAug
plt.plot(x, y1)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.scatter(x, y2)
plt.plot(x, y3)
plt.scatter(x, y3)
plt.plot(x, y4)
plt.scatter(x, y4)
plt.title('Classwise F1-Score Plot')
plt.ylabel('F1-Score')
plt.xlabel('Class')
plt.legend(['SequentialNetAug', 'ResidualNetAug', 'SqueezeNetAug', 'DepthwiseConvNetAug'], bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation = 60)
plt.ylim(bottom=0, top=100)
plt.show()
```
